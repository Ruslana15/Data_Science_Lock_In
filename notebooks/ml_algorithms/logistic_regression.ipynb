```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data = {
    'Площадь': [30, 40, 50, 60, 70, 80, 90, 100],
    'Купит':   [0, 0, 0, 0, 1, 1, 1, 1]  # 0 = нет, 1 = да
}

df = pd.DataFrame(data)
```

```python
X = df[['Площадь']]
y = df['Купит']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
model = LogisticRegression()

# обучение
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)

print("Предсказанные классы:", y_pred)
print("Реальные классы:", y_test.values)
print("Вероятности:\n", y_proba)
```

```python
plt.scatter(X, y, color='blue', label='Данные')

# линия вероятности
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_prob_line = model.predict_proba(X_range)[:, 1]

plt.plot(X_range, y_prob_line, color='red', label='Logistic Curve')

plt.xlabel('Площадь')
plt.ylabel('Вероятность покупки')
plt.legend()
plt.show()
```
